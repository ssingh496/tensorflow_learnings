```python
import tensorflow as tf
```

```python
a = tf.add(3,5)
sees = tf.Session()
print (sees.run(a))
print (sees.run(a))
sees.close()
```

```python
#use of with just to avoid the sees.close()
```

```python
a = tf.add(3,5)
with tf.Session() as sees:
    print(sees.run(a))
    
    
```

```python
#make your own graph
```

```python
g = tf.Graph()
with g.as_default():
    x = tf.add(3,5)
    
with tf.Session(graph=g) as sees:
    print(sees.run(x))
```

```python
# if you are making your own graph, it better to make it default or else it would be mix up with the tensorflow graph
```

```python
import tensorflow as tf
print (tf.__version__)
```

```python
a = tf.constant(2)
b = tf.constant(3)
c = tf.add(a, b)

with tf.Session() as sees:
    writer = tf.summary.FileWriter('./graphs/test', sees.graph)
    print(sees.run(c))

writer.close()
```

```python
tf.executing_eagerly()
a = tf.constant(2, name = "var_1")
b = tf.constant(3, name = "var_3")
x = tf.add(a,b, name = "add")
print(x)
```

```python
a= tf.zeros([3, 4], tf.int32)
```

```python
print(a)
with tf.Session() as sees:
    print(sees.run(a))
```

```python
a = [[2, 3], [1, 2]]
b = tf.zeros_like(a)
```

```python
with tf.Session() as sees:
    print(sees.run(b))
```

```python
#tf.fill = fill the tensor with the value you want
```

```python
a = tf.fill([2,3], 8)
with tf.Session() as sees:
    print(sees.run(a))
```

```python

%matplotlib inline
import pylab
import numpy as np

# create some data using numpy. y = x * 0.1 + 0.3 + noise
x = np.random.rand(100).astype(np.float32)
noise = np.random.normal(scale=0.01, size=len(x))
y = x * 0.1 + 0.3 + noise

# plot it
pylab.plot(x, y, '.')
```
